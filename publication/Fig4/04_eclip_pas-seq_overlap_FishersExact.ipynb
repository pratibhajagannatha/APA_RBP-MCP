```python
pas_candidate_paths = [i for i in glob.glob(os.path.join(pas_paths, '*.txt')) if i.split('/')[-1].split('_')[1] in candidates]
```

```python
def return_class(df):
    if df['log(treatment.p/treatment.d)'] > df['log(Ctrl.p/Ctrl.d)']:
        return 'proximal'
    elif df['log(treatment.p/treatment.d)'] < df['log(Ctrl.p/Ctrl.d)']:
        return 'distal'

modulated_sites = {}
totals = []
all_table_info = pd.DataFrame()
for rbp_table in pas_candidate_paths:
    kd = rbp_table.split('/')[-1].split('_')[1]
    modulated_sites[kd] = []
    test = pd.read_csv(rbp_table, sep='\t')
    test['treatment.p + 1'] = test['treatment.p'] + 1
    test['treatment.d + 1'] = test['treatment.d'] + 1
    test['Ctrl.p + 1'] = test['Ctrl.p'] + 1
    test['Ctrl.d + 1'] = test['Ctrl.d'] + 1
    test['log(treatment.p/treatment.d)'] = test.apply(lambda x: np.log2(x['treatment.p + 1']/x['treatment.d + 1']), axis=1)
    test['log(Ctrl.p/Ctrl.d)'] = test.apply(lambda x: np.log2(x['Ctrl.p + 1']/x['Ctrl.d + 1']), axis=1)
    test['apa_class'] = test.apply(return_class, axis=1)
    test['apa_type'] = test.apply(lambda x: f"{x['TYPE.p']}_{x['TYPE.d']}", axis=1)
    test['KD'] = kd
    test = test[(test['TYPE.p'].isin(['Intron', 'LastExon'])) & (test['TYPE.d'].isin(['Intron', 'LastExon']))]
    modulated_sites[kd] += test[test['apa_class']=='proximal']['COORDS.d'].tolist()
    modulated_sites[kd] += test[test['apa_class']=='distal']['COORD.p'].tolist()
    #hg38_coord = pd.read_csv(f'/tscc/nfs/home/pjaganna1/projects/apa_screen_yongsheng/apa_kd_all/pas-seq/hg38/{kd}_pas_hg38.bed', sep='\t', header=None)
    final_table = test[(test['KD'] ==kd)][['SYMBOL', 'COORD.p', 'COORDS.d', 'KD', 'apa_class']]
    all_table_info = pd.concat([all_table_info, final_table])
```

```python
from scipy.stats import fisher_exact
from statsmodels.stats.contingency_tables import Table2x2

def get_re_windows(prefix: str, skipper_output_dir: Path) -> pd.DataFrame:
    """
    Load reproducible enriched eCLIP windows and keep UTR3/INTRON features.
    Returns a DataFrame with columns: chrom,start,end,name,score,strand
    """
    window_path = skipper_output_dir / 'reproducible_enriched_windows' / f'{prefix}.reproducible_enriched_windows.tsv.gz'
    w = pd.read_csv(window_path, sep='\t')
    #w = w[w['feature_types'].str.contains('UTR3|INTRON', regex=True, na=False)].copy()
    #w = w.rename(columns={'chr': 'chrom'})
    cols = ['chr','start','end','name','score','strand']
    return w[cols].copy()

def bed_from_pas_df(l: int, df: pd.DataFrame, coord_mode: str) -> pybedtools.BedTool:
    """
    Build ±200bp windows around PAS coordinates.
    coord_mode ∈ {'COORD.p','COORDS.d','unfiltered'} chooses source column.
    Returns a BedTool (can be empty).
    """
    
    if df is None or df.empty:
        return pybedtools.BedTool("")

    signal_coord = []
    col = {'COORD.p': 'COORD.p', 'COORDS.d': 'COORDS.d', 'unfiltered': 'fPAS_ID'}
    for index, row in df.iterrows():
        #print(row[col[coord_mode]])
        ch, start,strand =  row[col[coord_mode]].split(':')[0], row[col[coord_mode]].split(':')[1], row[col[coord_mode]].split(':')[2]
        signal_coord.append([ch,int(start)-l,int(start) + l,
                                 row[col[coord_mode]], '.', strand])
            
    polysignal_df = pd.DataFrame(signal_coord,
                                columns= ['chr', 'start', 'end', 'name', 'score', 'strand']).drop_duplicates()

    return polysignal_df

def count_bound_unbound(pas_bed: pybedtools.BedTool, eclip_bed: pybedtools.BedTool, stranded: bool = True):
    """
    Count PAS windows that overlap eCLIP windows (bound) vs not (unbound).
    Uses -u to return only A intervals that overlap → fast & dedup.
    """
    if len(pas_bed) == 0:
        return 0, 0
    inter = pas_bed.intersect(eclip_bed,s = True, u= True).to_dataframe(header = None)
    bound = len(inter)
    total = len(pas_bed)
    return bound, total - bound

def fisher_or(table_2x2):
    """
    Compute OR, 95% CI, and two-sided Fisher p.
    Applies Haldane–Anscombe correction if needed.
    """
    a,b = table_2x2[0]
    c,d = table_2x2[1]
    OR, p = fisher_exact([[a,b],[c,d]])
    try:
        ci_lo, ci_hi = Table2x2([[a,b],[c,d]]).oddsratio_confint()
    except Exception:
        a+=0.5; b+=0.5; c+=0.5; d+=0.5
        OR, p = fisher_exact([[a,b],[c,d]])
        ci_lo, ci_hi = Table2x2([[a,b],[c,d]]).oddsratio_confint()
    return OR, (ci_lo, ci_hi), p

def eclip_bed_for_candidate(c: str) -> pybedtools.BedTool:
    """
    Get BedTool of reproducible eCLIP windows for candidate c.
    NUDT21 maps to CPSF5 eCLIP.
    """
    if c == 'NUDT21':
        pfx = 'CPSF5'
        base = Path('/tscc/nfs/home/pjaganna1/projects/apa_screen_yongsheng/eclip/skipper_runs/CPSF5/output/')
    else:
        pfx = c
        base = Path(f'/tscc/nfs/home/pjaganna1/projects/apa_screen_yongsheng/eclip/skipper_runs/{c}/output/')
    w = get_re_windows(pfx, base)
    #print(w)
    return w

def union_pas_bed(bt1: pd.DataFrame , bt2: pd.DataFrame) -> pybedtools.BedTool:
    """Exact-union two PAS sets (no merging of nearby windows)."""
    if bt1.shape[0] == 0 and bt2.shape[0] == 0:
        return pybedtools.BedTool("")

    df = pd.concat([bt1, bt2]).drop_duplicates(subset=['chr','start','end','name','strand'])
    #print(df)
    return df

# =============================================================================
# MAIN (requires these objects exist in your session):
#   - candidates: iterable of protein names
#   - all_table_info_dist: DataFrame with KD, COORDS.p (distal canonical), SYMBOL, etc.
#   - all_table_info_prox: DataFrame with KD, COORDS.d (proximal canonical), SYMBOL, etc.
#   - modulated_sites: dict[protein] -> set/list of fPAS_ID that are modulated (to exclude from unaffected)
# =============================================================================

rows = []
for c in candidates:
    print(c)
    # 1) eCLIP windows for this protein
    eclip_bed = eclip_bed_for_candidate(c)

    # 2) MODULATED PAS = PROXIMAL ∪ DISTAL (dedup)
    #    DISTAL canonical PAS from COORDS.p
    g_dist = all_table_info_dist[all_table_info_dist['KD'] == c]
    #print(g_dist.shape[0])
    pas_dist_bed = bed_from_pas_df(200, g_dist, 'COORD.p')

    #    PROXIMAL canonical PAS from COORDS.d
    g_prox = all_table_info_prox[all_table_info_prox['KD'] == c]
    #print(g_prox.shape[0])
    pas_prox_bed = bed_from_pas_df(200, g_prox, 'COORDS.d')

    pas_mod_bed = union_pas_bed(pas_prox_bed, pas_dist_bed)
    #print(pas_mod_bed)

    # 3) UNAFFECTED PAS (LastExon/Intron, excluding modulated fPAS_IDs)
    tag = 'CPSF5' if c == 'NUDT21' else c
    uf = pd.read_csv(f'./pas_not_sig/20241115_{tag}_APA_unfilterted.txt', sep='\t')
    uf = uf[uf['PAS_type.x'].isin(['LastExon','Intron'])].copy()
    #uf = uf[(~uf['fPAS_ID'].isin(modulated_sites[c])) & (uf['FDR'] >0.05) & (np.abs(uf['Change'])<0.15)].copy()
    uf = uf[~uf['fPAS_ID'].isin(modulated_sites[c])].copy()
    # This removes sites with low counts in ctr and treat
    uf_nosig =uf.sort_values(by=['ctr', 'treat'], ascending=[False, False]).copy()
    num_keep = int(uf_nosig.shape[0]*0.90)
    uf_nosig_keep = uf_nosig.head(num_keep)
    pas_uf_bed = bed_from_pas_df(200, uf_nosig_keep, 'unfiltered')
    #pas_uf_bed = bed_from_pas_df(200, uf, 'unfiltered')
    #print(pas_uf_bed.shape[0])
    #pas_uf_bed = pybedtools.BedTool.from_dataframe(pas_uf_bed)

    # 4) Count bound/unbound
    a_bound_mod, b_unbound_mod       = count_bound_unbound(pybedtools.BedTool.from_dataframe(pas_mod_bed), 
                                                           pybedtools.BedTool.from_dataframe(eclip_bed), stranded=True)
    c_bound_unaff, d_unbound_unaff   = count_bound_unbound(pybedtools.BedTool.from_dataframe(pas_uf_bed),  
                                                           pybedtools.BedTool.from_dataframe(eclip_bed), stranded=True)

    # 5) Compute OR vs UNAFFECTED
    if (a_bound_mod + b_unbound_mod == 0) or (c_bound_unaff + d_unbound_unaff == 0):
        OR = ci_lo = ci_hi = p = np.nan
    else:
        OR, (ci_lo, ci_hi), p = fisher_or([[a_bound_mod, b_unbound_mod],
                                           [c_bound_unaff, d_unbound_unaff]])

    rows.append({
        'protein': c,
        'comparison': 'MODULATED (prox∪dist) vs UNAFFECTED',
        'a_bound_mod': a_bound_mod,
        'b_unbound_mod': b_unbound_mod,
        'c_bound_unaff': c_bound_unaff,
        'd_unbound_unaff': d_unbound_unaff,
        'OR': OR,
        'CI_lo': ci_lo,
        'CI_hi': ci_hi,
        'p_fisher': p
    })

or_df = pd.DataFrame(rows).sort_values('protein').reset_index(drop=True)

# Optional: save results
# or_df.to_csv("odds_ratio_per_protein_modulated_vs_unaffected.tsv", sep="\t", index=False)

# Optional: quick peek
# print(or_df)
```

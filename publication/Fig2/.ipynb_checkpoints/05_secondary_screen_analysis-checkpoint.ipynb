```python
def make_table(file_list, r):
    all_table = pd.DataFrame()
    for f in tqdm.tqdm(file_list):
        if os.stat(f).st_size != 0:
            #print('True')
            t = pd.read_csv(f, sep='\t', index_col=0)
            t['rep1_mn'] = t['rep1']/t.loc['NEG','rep1']
            t['rep2_mn'] = t['rep2']/t.loc['NEG','rep2']
            t['rep3_mn'] = t['rep3']/t.loc['NEG','rep3']
            t['RBPmean-NEGmean'] = t['mean']-t.loc['NEG','mean']
            t['|RBPmean-NEGmean|'] = abs(t['RBPmean-NEGmean'])
            t['mean_normalized'] = t['mean']/t.loc['NEG','mean']
            t['Experiment'] = f.split('/')[-1].split('.txt')[0]
            if r=='Downstream':
                t['reporter']='Downstream'#annotating whether it is upstream or downstream
            elif r=='Upstream':
                t['reporter']='Upstream'#annotating whether it is upstream or downstream
            all_table=pd.concat([all_table, t.reset_index()])
            #print(all_table.shape[0])
    return all_table
```

```python
def process_secondary_screen(df_re):
    '''process secondary screen data, only keeping the top isoform'''
    df_re['original_well'] = df_re.well_id.map(redo_original_well_match)
    df_re['key'] = df_re.original_well.map(redo_original_key_match)
    df_re['pvalue'] = df_re['pvalue'].apply(lambda x: float(f"{x:.2g}"))
    df_re = df_re[(df_re['pvalue'] <0.05)]
    df_re = df_re[(df_re['RBPmean-NEGmean'] >0)]
    # only keep the top isoform and only keep activators
    df_re = df_re[(df_re['well_id']!='controls')].sort_values(by=['mean_normalized', 'pvalue'], ascending=[False, True]).drop_duplicates('index').copy()
    
    return df_re
```

```python
def get_sequences(s):
    sequences = []
    for i in s:
        if i in gene_to_sequence_dict:
            sequences.append(gene_to_sequence_dict[i])
        else:
            print(f"{i} not in gene to sequence dictionary")
    return set(sequences)
```

```python
def get_genes(seqs):
    genes = []
    for s in seqs:
        if s in sequence_to_gene_dict:
            genes.append(sequence_to_gene_dict[s])
        else:
            print(f"{s} not in dictionary")
    return set(genes)
```

```python
def plot_ranking_comparisons(s1, s2, dfs1, dfs2, figname, bins=10):
    # Prepare data
    s1_genes = get_genes(s1) # Original
    s2_genes = get_genes(s2) # Redo

    common_genes= set(list(s1_genes.intersection(s2_genes)))  
    for y in sorted(common_genes):
        print(y)
    unique_orig_only = list(s1_genes - common_genes)
    unique_redo_only = list(s2_genes - common_genes)
    #print(common_genes)
    #print(len(common_genes.union(unique_orig_only)))
    
    
    # Extract rankings
    common_orig_ranks = []
    unique_orig_ranks = []
    common_redo_ranks = []
    unique_redo_ranks = []
    
    for d1 in dfs1:
        common_orig_ranks += [ranking_dictionary[d1][g] for g in common_genes if g in ranking_dictionary[d1]]
        unique_orig_ranks += [ranking_dictionary[d1][g] for g in unique_orig_only if g in ranking_dictionary[d1]]

    for d2 in dfs2:
        common_redo_ranks += [ranking_dictionary[d2][g] for g in common_genes if g in ranking_dictionary[d2]]
        unique_redo_ranks += [ranking_dictionary[d2][g] for g in unique_redo_only if g in ranking_dictionary[d2]]
    print(len(common_orig_ranks),
         len(unique_orig_ranks),
         len(common_redo_ranks),
          len(unique_redo_ranks))
    
    # Plot
    fig, axs = plt.subplots(2, 1, figsize=(3, 5), sharex=True)
    
    # Define colors consistent with Venn diagram
    original_color = '#869A46'  # olive green
    redo_color = '#668EB1'      # muted blue
    
    # Common Genes Plot
    sns.kdeplot(common_orig_ranks, color=original_color, label='Primary Screen Rank',
               ax=axs[0])
    #sns.kdeplot(common_redo_ranks, color=redo_color, label='Secondary Screen Rank', ax=axs[0])
    axs[0].set_title('Ranking Distribution of Common Proteins')
    axs[0].legend()
    
    # Unique Genes Plot
    sns.kdeplot(unique_orig_ranks, color=original_color, label='Unique to Primary Screen', ax=axs[1])
    #sns.kdeplot(unique_redo_ranks, color=redo_color, label='Unique to Secondary Screen', ax=axs[1])
    axs[1].set_title('Ranking Distribution of Unique Proteins')
    axs[1].legend()
    
    for ax in axs:
        ax.set_ylabel('Density')
        ax.set_xlabel('Rank')
    
    plt.tight_layout()
    # plt.savefig(f'./figures/{figname}.png')
    # plt.savefig(f'./figures/{figname}.svg')
    # plt.savefig(f'./figures/{figname}.jpg')
    plt.show()
```
